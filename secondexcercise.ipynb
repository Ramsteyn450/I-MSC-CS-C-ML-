```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.preprocessing import StandardScaler


x = np.array([1, 5, 7, 8, 9])
y = np.array([2, 6, 9, 10, 12])   # <-- You MUST define y (added example values)

print(x.shape)
print(y.shape)
print(x)
print(y)

# reshape x for sklearn
x = x.reshape(-1, 1)
print(x.shape)
print(x)

model = LinearRegression()
model.fit(x, y)

print(model.predict([[8]]))

y_pred = model.predict(x)

plt.scatter(x, y, color="r", marker="o")
plt.plot(x, y_pred, color="g", linewidth=2)
plt.xlabel("x da")
plt.ylabel("y da")
plt.show()

print(model.coef_)
print(model.intercept_)


scaler = StandardScaler()
x_stand = scaler.fit_transform(x.astype(float))

sgdr = SGDRegressor(penalty="l2", alpha=0.15, max_iter=1000)
sgdr.fit(x_stand, y)

print(sgdr.predict(scaler.transform([[10]])))

y_pred2 = sgdr.predict(x_stand)

plt.scatter(x_stand, y, color="y", marker="o")
plt.plot(x_stand, y_pred2, color="b", linewidth=2)
plt.xlabel("x_normalized")
plt.ylabel("y")
plt.show()

print(sgdr.coef_)
print(sgdr.intercept_)
```
